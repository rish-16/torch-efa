```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
```

```python
# Define grid
theta = np.linspace(0, np.pi, 100)    # polar angle
phi = np.linspace(0, 2*np.pi, 100)    # azimuthal angle
theta, phi = np.meshgrid(theta, phi)

# Spherical harmonic parameters
l = 3  # degree
m = 1  # order
```

```python
# Compute complex spherical harmonics
Y_lm = sph_harm(m, l, phi, theta)

# Convert to Cartesian coordinates for plotting
r = np.abs(Y_lm.real)  # or Y_lm.imag or np.abs(Y_lm)
x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)
```

```python
# Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, facecolors=plt.cm.viridis((r - r.min())/(r.max() - r.min())))
ax.set_title(f"Real part of Y_{l}^{m}")
plt.show()
```
